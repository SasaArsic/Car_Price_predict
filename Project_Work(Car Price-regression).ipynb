```python
# Is there any NaN value?
# Let's use function to identify missing value from all features
def missing_values (dataset):
    total = df.isnull().sum().sort_values(ascending = False)
    percent = ((df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False))
    missing_values= pd.concat([total,percent], axis=1, keys= ['Total', 'Percent'])
    return missing_values

print('Missing Values from Car Price data set:')
print()
print (missing_values (df))
```

# Let's apply Machine Learning models

```python
# import mean absolute and squared error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Create dict acc_scores to see which model is the best
MSE= {}
MAE= {}
R2={}
model_list=[]
model_name=[]

# Import models that will be applied on dataset

# First model Linear Regression
from sklearn.linear_model import LinearRegression
Lin_reg= LinearRegression ()
model= Lin_reg
model_list.append(model)

# Second model SVC kernel linear
from sklearn.svm import SVR, LinearSVR
Svr_linear = SVR (kernel='linear', C=1)
model= Svr_linear
model_list.append(model)

# Third model SVR_rbf
SVR_rbf= SVR(kernel='rbf', C=100)
model= SVR_rbf
model_list.append(model)

# Forth model RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
Ran_for_reg = RandomForestRegressor(random_state=0)
model= Ran_for_reg
model_list.append(model)

# Fifth model KNN
from sklearn.neighbors import KNeighborsRegressor
KNN = KNeighborsRegressor()
model= KNN
model_list.append(model)

# 6th XGBoostRegressor
import xgboost as xgb
Xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10, random_state=0)
model= Xg_reg
model_list.append(model)

# 7th model DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
Dec_tree_reg = DecisionTreeRegressor(random_state=0)
model= Dec_tree_reg
model_list.append(model)

# 8th model BaggingRegressor
from sklearn.ensemble import BaggingRegressor
Bag_reg = BaggingRegressor(random_state=0)
model= Bag_reg
model_list.append(model)

#9th model RANSACRegressor
from sklearn.linear_model import RANSACRegressor
Ransac = RANSACRegressor(random_state=0)
model= Ransac
model_list.append(model)

#10th model lasso
from sklearn.linear_model import Lasso
Lasso = Lasso(random_state=0)
model= Lasso
model_list.append(model)
```

```python
#fuction model_func takes arguments( model,X_train,y_train,X_test,y_test)
# first  model is fitted with X_train,X_test
#next, prediction is made by model
#third errors are calculated,and appended in appropiate dict

def model_func(model,X_train,y_train,X_test,y_test):
    model=model
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    mae=mean_absolute_error(y_test,y_pred)
    mse=mean_squared_error(y_test,y_pred)
    r2=r2_score(y_pred, y_test)
    MAE[model]= mae
    MSE[model]= mse
    R2 [model]= r2
    
    return(model,MAE,MSE,y_pred,R2)
```

```python
def unique_list_model(list_model):
    model_unique=[]
    for model in list_model:
        if model not in model_unique:
            model_unique.append(model)
    return(model_unique)
```

```python
# Since citympg and highwaympg are so strongly correlated and when you think about it logically these 2 variables 
# are both related to the car's fuel economy and how many mpg the car can drive.
    
# I will combine these 2 variables into one by taking the adding the values and dividing by 2. 
# This will give a rough "average mpg".

df1["avgmpg"] = (df1["citympg"]+df1["highwaympg"])/2
df1["avgmpg"].head(5)
```

```python
# Let's deal with the price range per manufacturer, we will divide the car into 4 categories, Cheap, Affordable, Expensive and Luxury

df1["CarBrand"] = df1['price'].apply(lambda x : "Cheap" if x < 10000 
                                                     else ("Affordable" if 10000 <= x < 15000
                                                           else ('Expensive' if 15000 <= x < 25000 
                                                           else ("Luxury"))))
```

```python
#Symboling refers to the category of insurance, let's put it that way and divide it into cars assessed at high risk, low risk and no risk

df1["symboling"]=df1["symboling"].apply(lambda x : "No risk" if x >= -3 & x <= -1
                                         else ("Low risk" if x>=0 and x <= 1
                                            else ("High risk")))
```

```python
# As the feature cylinder numer is essentially numerical, we will map it to the numerical values of its string meaning.
df1['cylindernumber'] = df1['cylindernumber'].map({'two':2, 'three': 3,'four': 4, 
                                                             'five': 5,'six': 6,'eight': 8,'twelve':12})
```

```python
# Next we will use Label Encoder on some of our categorical features
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
df1['fueltype'] = le.fit(df1['fueltype']).transform(df1['fueltype'])
df1['aspiration'] = le.fit(df1['aspiration']).transform(df1['aspiration'])
df1['doornumber'] = le.fit(df1['doornumber']).transform(df1['doornumber'])
df1['enginelocation'] = le.fit(df1['enginelocation']).transform(df1['enginelocation'])
```
